## Step 1

```python
board = Board()

# Place orange blooms
board.place_stone(((1, 4)), 1)
board.place_stone(((2, 6)), 1)

# Place green bloom
board.place_stone(((6, 2)), 2)
board.place_stone(((6, 3)), 2)
board.place_stone(((5, 4)), 2)

# Place red bloom
board.place_stone(((3, 1)), 3)
board.place_stone(((2, 2)), 3)

# Place purple bloom
board.place_stone(((6, 1)), 4)
board.place_stone(((5, 2)), 4)
board.place_stone(((5, 3)), 4)
board.place_stone(((4, 4)), 4)


board.visualise(show_coords=False, filename='outputs/gameplay_sequence_1.pdf')
```
