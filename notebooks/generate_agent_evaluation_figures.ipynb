# Generate the Plots for the Checkpoint Evaluation of AlphaBlooms Agents

## Imports

```python
import pickle
import sys
sys.path.append('..')

import numpy as np
import matplotlib.pyplot as plt

import Arena
from MCTS import MCTS
from utils import *

from blooms.BloomsGame import BloomsGame
from blooms.BloomsPlayers import RandomPlayer, GreedyPlayer
from blooms.pytorch.NNet import NNetWrapper as BloomsPyTorchNNet
```

## Base 4 Board

```python
CHKPT_DIR = 'results/chkpts_board4_24hrs/'  # The path to the checkpoints directory

results_vs_random = pickle.load(open(CHKPT_DIR + 'chkpt_evals_vs_random_agent.pkl', 'rb'))
results_vs_greedy = pickle.load(open(CHKPT_DIR + 'chkpt_evals_vs_greedy_agent.pkl', 'rb'))
```

```python
fig, ax = plt.subplots(1)


checkpoints = range(1, 13 + 1)
plt.plot(checkpoints, results_vs_random, label='Random')
plt.plot(checkpoints, results_vs_greedy, label='Greedy')

plt.xlabel('Iteration')
plt.ylabel('Win Proportion')
plt.ylim(0, 1)
plt.xticks(range(1, 14))
plt.legend()
plt.title('Progressive Performance Against the Random and Greedy\nAgent Baselines on a Base 4 Board')
plt.savefig('outputs/performance_over_time_base_4.pdf')
plt.show()
```
